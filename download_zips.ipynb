```python
quality = "scribble" # one of high, scribble and nolabel
```

```python
links = None
with open("./cloudSEN12_links.txt", "r") as f:
    lines = f.readlines()
    links = [l.rstrip() for l in lines]
```

```python
links
```

```python
selected_p509_links = [l for l in links if quality in l and "p509" in l]
```

```python
selected_p509_links
```

```python
len(selected_p509_links)
```

```python
import requests, io, zipfile

i = 17
for l in selected_p509_links[i-1:]:
    folder = l.split("/")[7]
    print(f"downloading {i}th link")
    r = requests.get(l)
    z = zipfile.ZipFile(io.BytesIO(r.content))
    z.extractall(f"./data/p509/{quality}/{folder}/")
    print(f"done {r.status_code}")
    with open(f"./data/p509/{quality}/{folder}_dl_progress.txt", "a+") as f:
        f.write(f"i{i} done ({l})\n") 
    i += 1
```
